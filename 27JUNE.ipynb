```python
fig = plt.figure()


with plt.style.context('Solarize_Light2'):
    fig.set_facecolor('white')
    plt.rcParams['axes.facecolor'] = 'white'
    plt.plot(bba_df_8['intSeg'], bba_df_8['rtt_qos'], label='BBA-2 +')
    plt.plot(logistic_df_8['intSeg'], logistic_df_8['rtt_qos'], label='Logistic',linestyle='--', color='orange')
    #plt.plot(x, np.sin(x) + 3 * x + np.random.randn(50))
    #plt.plot(x, np.sin(x) + 4 + np.random.randn(50))
    #plt.plot(x, np.sin(x) + 5 * x + np.random.randn(50))
    #plt.plot(x, np.sin(x) + 6 * x + np.random.randn(50))
    #plt.plot(x, np.sin(x) + 7 * x + np.random.randn(50))
    #plt.plot(x, np.sin(x) + 8 * x + np.random.randn(50))
    # Number of accent colors in the color scheme
    plt.title('Bandwidth: 0.5 - 4 Mbps', fontsize=12)
    plt.xlabel('Segments (2 sec)', fontsize=12, color='black')
    plt.ylabel('Per-segment RTT', fontsize=12, color='black')
    plt.legend(frameon=False)
    # Draw grid lines with red color and dashed style
    plt.grid(axis='y', c='grey',alpha=0.2)
    plt.grid(axis='x', c='grey',alpha=0.2)
    plt.tick_params(axis='x', colors='black')
    plt.tick_params(axis='y', colors='black')

plt.savefig('NEW/buffered_5_4MBPS.png', format='png')      
plt.show()
```

```python
fig, axs = plt.subplots(5,figsize=(6,5))
fig.suptitle('Bandwidth: 0.5 - 4 Mbps')

axs[0].plot(bba_df_8['intSeg'], bba_df_8['P1203'], label='P1203',linestyle='--')
axs[0].set(ylabel='P.1203')
axs[1].plot(bba_df_8['intSeg'], bba_df_8['Clae'], label='Clae',linestyle='--')
axs[1].set(ylabel='Clae')
axs[2].plot(bba_df_8['intSeg'], bba_df_8['Duanmu'], label='Duanmu',linestyle='--')
axs[2].set(ylabel='Duanmu')
axs[3].plot(bba_df_8['intSeg'], bba_df_8['Yin'], label='Yin',linestyle='--')
axs[3].set(ylabel='Yin')
axs[4].plot(bba_df_8['intSeg'], bba_df_8['Yu'], label='Yu',linestyle='--')
axs[4].set(ylabel='Yu')
for ax in axs.flat:
    ax.label_outer()
    
plt.savefig('NEW/buffered_all_qoe_5_4MBPS.png', format='png')
```
